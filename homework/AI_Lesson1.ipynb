```python
import pandas as pd
import os

CSV_PATH = os.path.join("datasets", "athlete_events.csv")
data = pd.read_csv(CSV_PATH)
data.head()
```

```python
# 1. How old were the youngest male and female participants of the 1996 Olympics?
data[(data['Sex']=='M') & (data['Year']==1996)]['Age'].min()
```

```python
data[(data['Sex']=='F') & (data['Year']==1996)]['Age'].min()
```

```python
# 2. What was the percentage of male gymnasts among all the male participants of the 2000 Olympics? 
# Round the answer to the first decimal.

# Hint: here and further if needed drop duplicated sportsmen to count only unique ones.

100 * len(data[(data['Sport']=="Gymnastics") & (data['Sex']=='M')  & (data['Year']==2000)]['Name'].drop_duplicates()) / len(data[(data['Sex']=='M')  & (data['Year']==2000)]['Name'].drop_duplicates())
```

```python
# 3. What are the mean and standard deviation of height for female basketball players participated in the 2000 Olympics? 
# Round the answer to the first decimal.

data[(data['Sex']=='F') & (data['Year']==2000) & (data['Sport']== 'Basketball')]['Height'].describe()
```

```python
# 4. Find a sportsperson participated in the 2002 Olympics, 
# with the highest weight among other participants of the same Olympics. 
# What sport did he or she do?

data[(data['Year']==2002)].sort_values(by='Weight', ascending=False).head()
```

```python
# 5. How many times did Pawe Abratkiewicz participate in the Olympics held in different years?
data[data['Name'] == 'Pawe Abratkiewicz']
```

```python
# 6. How many silver medals in tennis did Australia win at the 2000 Olympics?

data[(data['Year']==2000) & (data['Team']=='Australia') & (data['Sport']=='Tennis') ] 
```

```python
# 7. Is it true that Switzerland won fewer medals than Serbia at the 2016 Olympics? Do not consider NaN values in Medal column.

data[(data['Year']==2016) & (data['Team']=='Switzerland')]['Medal'].value_counts() 
```

```python
data[(data['Year']==2016) & (data['Team']=='Serbia')]['Medal'].value_counts() 
```

```python
#  What age category did the fewest and the most participants of the 2014 Olympics belong to?
data[(data['Year']==2014) & (data['Age']<45) & (data['Age']>=35)].drop_duplicates(['Name'])['Name'].count()


```

```python
# Is it true that there were Summer Olympics held in Lake Placid? 
# Is it true that there were Winter Olympics held in Sankt Moritz?

data[(data['City'] == 'Sankt Moritz') & (data['Season'] == 'Winter')]
```

```python
# 10. What is the absolute difference between the number of unique sports at the 1995 Olympics and 2016 Olympics?

data[(data['Year']==2016)]['Sport'].count()
```
